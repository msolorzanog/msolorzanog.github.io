```
#Librerias 
import numpy as np
import pandas as pd
import cv2  
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
```

Actividad 1. Aplicar Filtro adecuado - Segmentar Tejido 

```
url2="https://static-01.hindawi.com/articles/cmmm/volume-2014/690349/figures/690349.fig.006e.jpg"
image1 = io.imread(url2) 
cv2_imshow(image1) 
median=cv2.medianBlur (image1, 3)
cv2_imshow(median) 
```

```
img=cv2.cvtColor(median,cv2.COLOR_BGR2RGB)
vectorized = img.reshape((-1,3))
vectorized = np.float32(vectorized)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 2
attempts=10
ret,label,center=cv2.kmeans(vectorized,K,None,criteria,attempts,cv2.KMEANS_PP_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
result_image = res.reshape((img.shape))
cv2_imshow(result_image)
```
